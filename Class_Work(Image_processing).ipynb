```python
import numpy as np
def apply_juno_filter(image):
    # Apply color enhancement
    color_enhanced = cv2.applyColorMap(image, cv2.COLORMAP_AUTUMN)
    # Increase contrast and brightness
    alpha = 1.2  # Contrast control (1.0 for no change)
    beta = 10    # Brightness control (0 for no change)
    juno_filtered = cv2.convertScaleAbs(color_enhanced, alpha=alpha, beta=beta)
    return juno_filtered
# Load the original image
image_path = "WIN_20230722_08_20_25_Pro.jpg"  # Replace with the path to your image
original_image = cv2.imread(image_path)
# Convert BGR image to RGB (OpenCV loads images in BGR format)
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
# Display the original image
plt.subplot(1, 2, 1)
plt.imshow(original_image_rgb)
plt.title("Original Image")
plt.axis("off")
# Apply Juno filter
juno_filtered_image = apply_juno_filter(original_image)
# Convert BGR image to RGB for display
juno_filtered_image_rgb = cv2.cvtColor(juno_filtered_image, cv2.COLOR_BGR2RGB)
# Display the Juno filtered image
plt.subplot(1, 2, 2)
plt.imshow(juno_filtered_image_rgb)
plt.title("Juno Filtered Image")
plt.axis("off")
plt.show()
```

```python
def pencil_sketch(image_path):
    # Load the image
    image = cv2.imread("WIN_20230722_08_20_25_Pro.jpg")
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Invert the grayscale image
    inverted_image = 255 - gray_image
    # Apply Gaussian blur to the inverted image
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), sigmaX=0, sigmaY=0)
    # Invert the blurred image
    inverted_blurred_image = 255 - blurred_image
    # Create the pencil sketch image by blending the grayscale and inverted blurred images
    pencil_sketch_image = cv2.divide(gray_image, inverted_blurred_image, scale=256.0)
    # Convert BGR image to RGB for display
    pencil_sketch_image_rgb = cv2.cvtColor(pencil_sketch_image, cv2.COLOR_BGR2RGB)
    # Display the original image and the pencil sketch image
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(image)
    plt.title("Original Image")
    plt.axis("off")
    plt.subplot(1, 2, 2)
    plt.imshow(pencil_sketch_image_rgb)
    plt.title("Pencil Sketch Image")
    plt.axis("off")
    plt.show()
if __name__ == "__main__":
    image_path = "WIN_20230722_08_20_25_Pro.jpg"  # Replace with the path to your image
    pencil_sketch(image_path)
```

```python
#import the packages
from skimage import io, color
from skimage.filters import sobel
from skimage.segmentation import felzenszwalb
import matplotlib.pyplot as plt
# Load the image
img = io.imread("WIN_20230722_08_20_25_Pro.jpg")
# Convert to grayscale
grayimg = color.rgb2gray(img)
# Apply the felzenszwalb segmentation
segments_fz = felzenszwalb(img, scale=100, sigma=0.5, min_size=50)
# Convert to a cartoon version
cartoon = color.label2rgb(segments_fz, img, kind='avg')
#Display original image
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
#Display cartoon image
plt.figure(figsize=(5,5))
plt.imshow(cartoon)
plt.axis("off")
plt.title("Cartoon Image")
plt.show()
```

```python
from PIL import Image
# Load the image
image_path = "WIN_20230722_08_20_25_Pro.jpg"  # Replace with the path to your image
image = Image.open(image_path)
from PIL import ImageEnhance, ImageFilter
# Convert the image to grayscale
grayscale_image = image.convert("L")
# Apply the Willow filter
willow_filter = ImageEnhance.Color(grayscale_image).enhance(0.5)
willow_filter = willow_filter.filter(ImageFilter.GaussianBlur(radius=10))
# Display the filtered image
plt.imshow(willow_filter)
plt.title("Willow Filtered Image")
plt.axis("off")
plt.show()
```
